# Signals

```
import numpy as np
import matplotlib.pyplot as plt
```

```
t = np.linspace(0, 2.0, 80)
```

$$x(t) = A\sin(2\pi f t + phase)$$

```
def create_sinusoidal(A, f, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

**Create a Sinusoidal Signal $x(t)$ and 3 different plotting in subplots**

```
x = create_sinusoidal(1,1)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('Different subplots of a sinusoidal signal', fontsize=15)

ax[0, 0].plot(t,x)
ax[0, 0].set_title('continuous plot', fontsize = 15)

ax[0, 1].stem(t,x)
ax[0, 1].set_title('stem plot', fontsize = 15)

ax[1,0].scatter(t,x)
ax[1,0].set_title('scatter plot', fontsize = 15)
ax[-1,-1].axis('off')

# for a in ax.flat:
#     a.set(xlabel='t', ylabel='x(t)')

plt.show()
```

**Create 3 Sinusoidal Signal $x1(t),x2(t),x3(t)$ then combine and plot the signal**

```
x1 = create_sinusoidal(1,1)
x2 = create_sinusoidal(2,5)
x3 = create_sinusoidal(3,3)
```

```
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('3 Sinusoidal signal and a combined signal', fontsize=15)

ax[0, 0].plot(t,x1, 'r')
ax[0, 0].set_title('x1(t)', fontsize = 15)

ax[0, 1].plot(t,x2, 'g')
ax[0, 1].set_title('x2(t)', fontsize = 15)

ax[1,0].plot(t,x3, 'b')
ax[1,0].set_title('x3(t)', fontsize = 15)

ax[1,1].plot(t, x1 + x2 + x3)
ax[1,1].set_title('y(t) = x1(t) + x2(t) + x3(t)', fontsize = 15)
plt.show()
```

```
t = np.arange(0.0,2.0,0.00001)
```

$$x(t) = A\sin(2\pi f t + phase)$$

```
def create_sinusoidal(t, A, f, phase = 0):
  x = A * np.sin(2.0 * np.pi* f * t + phase)
  return x
```

```
x1 = create_sinusoidal(t, 1, 5)
x2 = create_sinusoidal(t, 1, 10)
x3 = create_sinusoidal(t, 1, 15)
y = x1 + x2 + x3
```

```
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('3 Sinusoidal Signals and 1 Composite Signal', fontsize=15)
ax[0,0].plot(t, x1)
ax[0,0].set_title('Signal-1', fontsize = 15)

ax[0,1].plot(t, x2)
ax[0,1].set_title('Signal-2', fontsize = 15)

ax[1,0].plot(t, x3)
ax[1,0].set_title('Signal-3', fontsize = 15)

ax[1,1].plot(t, y)
ax[1,1].set_title('Signal-1 + Signal-2 + Signal-3', fontsize = 15)

plt.show()
```

```
fs =  10
step  = 1.0/fs
n  = np.arange(0, 2.0, step)
y  = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs  = 10', fontsize  = 15)
plt.stem(n,y)
plt.show()

fs  = 20
step  = 1.0/fs
n  = np.arange(0, 2.0, step)
y  = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs  = 20', fontsize  = 15)
plt.stem(n,y)
plt.show()

fs  = 50
step  = 1.0/fs
n  = np.arange(0, 2.0, step)
y  = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs  = 50', fontsize  = 15)
plt.stem(n,y)

plt.show()
```
